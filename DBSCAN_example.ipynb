```python
#let's implement some dbscan
from sklearn.cluster import DBSCAN
```

```python
#let's create our feature
"""
potential_features
salary_currency
salary
remote ratio
job_title
experience level
company size
"""
```

```python
df.info()
```

```python
feature=df.iloc[:,[1,3,5,8,10,11]]
```

```python
feature=feature.drop_duplicates()
```

```python
feat=pd.get_dummies(feature)
```

```python
feat=feat.drop(columns='salary_in_inr')
```

```python
cluster=DBSCAN(eps=8,min_samples=9).fit(feat)
```

```python
cluster.labels_
```

```python
feature['cluster_labels']=cluster.labels_
```

```python
feature
```

```python
feature.cluster_labels.unique()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(feat,cluster.labels_)#assigned to wrong clusters and overlapping
```
